```python
# 2. Memuat dataset
from google.colab import files
uploaded = files.upload()  # Upload file ARTO.csv
```

```python
data = pd.read_csv('ARTO.csv')
```

```python
# 3. Membersihkan dan memisahkan kolom
data_cleaned = data['timestamp;open;low;high;close;volume'].str.split(';', expand=True)
data_cleaned.columns = ['timestamp', 'open', 'low', 'high', 'close', 'volume']
```

```python
# Konversi kolom numerik ke tipe data yang benar
data_cleaned['open'] = pd.to_numeric(data_cleaned['open'], errors='coerce')
data_cleaned['low'] = pd.to_numeric(data_cleaned['low'], errors='coerce')
data_cleaned['high'] = pd.to_numeric(data_cleaned['high'], errors='coerce')
data_cleaned['close'] = pd.to_numeric(data_cleaned['close'], errors='coerce')
data_cleaned['volume'] = pd.to_numeric(data_cleaned['volume'], errors='coerce')
```

```python
# 4. Menyiapkan data untuk model
# Menggunakan fitur 'open', 'low', 'high', dan 'volume' untuk memprediksi 'close'
X = data_cleaned[['open', 'low', 'high', 'volume']]
y = data_cleaned['close']
```

```python
# Membagi dataset menjadi training set dan test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 5. Membuat model regresi linier
model = LinearRegression()
```

```python
# 6. Melatih model
model.fit(X_train, y_train)
```

```python
# 7. Memprediksi hasil
y_pred = model.predict(X_test)
```

```python
# 8. Evaluasi model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error: {rmse}')
```

```python
# 9. Visualisasi hasil
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Close Prices')
plt.ylabel('Predicted Close Prices')
plt.title('Actual vs Predicted Close Prices')
plt.show()
```
